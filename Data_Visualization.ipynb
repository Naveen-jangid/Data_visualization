#### Que:- 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
size = 100  # Number of samples
X1 = np.random.normal(50, 10, size)  # Feature 1
X2 = X1 * 0.8 + np.random.normal(0, 5, size)  # Feature 2 (correlated with X1)
X3 = X1 * -0.6 + np.random.normal(0, 7, size)  # Feature 3 (negatively correlated with X1)
X4 = np.random.normal(30, 8, size)  # Feature 4 (less correlation)

# Create DataFrame
df = pd.DataFrame({"Feature 1": X1, "Feature 2": X2, "Feature 3": X3, "Feature 4": X4})

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#### Que:-3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
# Set random seed
np.random.seed(20)

# Generate sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Create a pivot table for heatmap
pivot_df = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create heatmap using Plotly
fig = px.imshow(pivot_df, 
                labels=dict(x="Month", y="Day", color="Sales"), 
                x=pivot_df.columns, 
                y=pivot_df.index, 
                color_continuous_scale="viridis",
                title="Sales Variation Across Months and Days")

# Show plot
fig.show()
```

#### Que:-4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x^2+y^2))

```python
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z values based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale="viridis")])

# Update layout for better visualization
fig.update_layout(title="3D Surface Plot of z = sin(√(x² + y²))",
                  scene=dict(xaxis_title="X",
                             yaxis_title="Y",
                             zaxis_title="Z"))

# Show the plot
fig.show()
```

#### Que:-2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Enable Bokeh to display in a notebook
output_notebook()

# Generate random data
np.random.seed(42)
num_points = 100
x = np.random.rand(num_points) * 10  # X values between 0 and 10
y = np.random.rand(num_points) * 10  # Y values between 0 and 10
sizes = np.random.randint(5, 30, num_points)  # Marker sizes
colors = np.random.randint(0, 256, num_points)  # Colors mapped to a colormap

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot with Variable Sizes and Colors",
           x_axis_label="X", y_axis_label="Y", width=700, height=400)

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=ColumnDataSource(df),
          fill_color="firebrick", line_color="black")

# Show plot
show(p)
```

#### Que:3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

#### Que:- 4. Ceate a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.models import ColumnDataSource

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create figure
p = figure(title="Histogram of Data Distribution",
           x_axis_label="Value", y_axis_label="Frequency",
           width=600, height=400)

# Add histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show plot
show(p)
```

#### Que:- 5. Create Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform

# Generate Data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create ColumnDataSource
source = ColumnDataSource(data=dict(x=xx.flatten(), y=yy.flatten(), value=data_heatmap.flatten()))

# Define Color Mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create Figure
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y",
           width=600, height=400)

# Add Heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=transform('value', color_mapper), line_color=None)

# Add Color Bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
